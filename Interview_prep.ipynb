```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/employee_data.csv',header=0,\
                 names=['id','first_name','status','sex','birth_date','country','hire_date','terminate_date','aa','bb','cc'])
```

```python
df['id']= df['id'].astype(int)
df['first_name']= df['first_name'].astype(str)
df['hire_date']= pd.to_datetime(df['hire_date'], infer_datetime_format=True)
df['terminate_date']= pd.to_datetime(df['terminate_date'], infer_datetime_format=True)
```

```python
df_hire = pd.DataFrame(df['hire_date'].rename("date"))

df_hire['value']=1
```

```python
df_terminate = pd.DataFrame(df['terminate_date'].rename("date")).dropna()

df_terminate['value']=-1
```

```python
all_dates = pd.concat([df_hire,df_terminate], ignore_index=True)
```

```python
all_df = all_dates.groupby("date").sum().reset_index()
```

```python
all_df["employee_cnt"] = all_df.sort_values("date")['value'].cumsum()
```

```python
all_df.sort_values("date")
```

```python
from datetime import date

df["terminate_date_new"] = df["terminate_date"].mask(pd.isnull, date.today().strftime("%d-%m-%Y"))
```

Using Another approach using just python code

```python
for i in range(df.shape[0]): 
  start_date = df.at[i,"hire_date"]
  end_date = df.at[i,"terminate_date_new"]
  max_employee_cnt= all_df[all_df["date"].between(start_date, end_date)]["employee_cnt"].max()
  earliest_date = all_df[ (all_df["date"].between(start_date, end_date)) & (all_df["employee_cnt"]==max_employee_cnt)]["date"].min()
  df.at[i,"min_date"] = earliest_date
  df.at[i, "max_employee_cnt"] = max_employee_cnt
```
